A jupyter notebook works as a client-server application. The server is usually started on your local system in a Terminal.

```
$ jupyter-notebook [list]
```

Options:

* ``--log-level=<Enum>``
* ``--notebook-dir=<Folder>``

The log-level above can be an integer or one of (DEBUG=0, INFO=10, WARNING=20, **ERROR=30**, CRITICAL=40). The default is 30.

## Working with the Jupyter Notebook

Using the command pallet to learn keyboard shortcuts: ``⇧`` + ``⌘`` + ``P``


## Command & Edit Mode

The ``esc`` key will bring you into **command mode** where you basically manipulate and run cells and interact with the Jupyter Notebook itself.

The ``Return`` key will bring you (back) into edit mode where you edit your code or write markdown to document your notebook.

## Show documentation

``⇧`` + ``TAB`` shows the documentation or docstring for the object you have just typed. Pressing ``⇧`` + ``TAB`` again will toggle you through different docstring options.

Alternatively, you can add a ``?`` to the end of the object name and render the cell with ``⇧`` + ``Return``.

````python
import numpy as np

x = np.zeros((2,3))
x
````

````python
x?
````

## Manipulating cells

| Key code | Action |
|   :---:  | :---   |
| ``⇧`` + ``Down`` | Select the next cell downwards |
| ``⇧`` + ``Up`` | Select the next cell upwards |
| ``a``    | Add a new cells above |
| ``b``    | Add a new cell below |
| ``x``    | Cut the current cell or selected cells |
| ``c``    | Copy the current cell or selected cells |
| ``v``    | Paste the 
| ``^`` + ``⇧`` + ``Minus``   | Split the cell at the cursor position |
| ``⇧`` + ``M``   | Merge current and next cell |
